# univariate Analysis

```python
## defining functions for univariate analysis
def non_viz_num(data):
    print(data.agg(["min", "max", "mean", "median","std", "skew", "kurt"]))
    
    
def viz_num(data):
    fig, ax = plt.subplots(1, 3, figsize=(10,5), constrained_layout=True)
    
    ax[0].set_title("Histogram Plot")
    sns.histplot(data, ax=ax[0])
    
    ax[1].set_title("KDE plot")
    sns.kdeplot(data, ax=ax[1])
    
    ax[2].set_title("Box plot")
    sns.boxplot(data, ax=ax[2])
    
    plt.show()
    
    
def non_viz_cat(data):
    print(data.agg(["count", "nunique", "unique"]))
    print(data.value_counts())
    
def viz_cat(column_name):
    top_values = df[column_name].value_counts().head(10).index

    filtered_df = df[df[column_name].isin(top_values)]

    plt.figure(figsize=(10, 8))
    sns.countplot(x=column_name, data=filtered_df, order=top_values)
    
    plt.xticks(rotation=45, ha='right')

    plt.title('Top 10 Values Count Plot')

    plt.show()
```

## Removing outliers using quartile range


```python
for column in numerical_df.columns:
    Q1 = numerical_df[column].quantile(0.25)
    Q3 = numerical_df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    numerical_df = numerical_df[(df[column] >= lower_bound) & (numerical_df[column] <= upper_bound)]
```

## verifying using QQ plot


```python
def QQplot(data):
    stats.probplot(data, dist=stats.norm, plot=plt)
    plt.grid()

QQplot(numerical_df["Salary"])
```

### After viewing QQ plot, it is verified that Salary in not normally distributed


## Shapiro wilk test for Normality

```python
from scipy.stats import shapiro
     
statistic, p_value = shapiro(numerical_df['Salary'])

alpha = 0.05
if p_value > alpha:
    print("Salary is normally distributed (fail to reject the null hypothesis)")
else:
    print("Salary is not normally distributed (reject the null hypothesis)")
```

## Kolmogorov-Smirnov (K-S) test for distribution

```python
from scipy.stats import ks_2samp

salary = numerical_df['Salary']

theoretical_distribution = np.random.normal(loc=salary.mean(), scale=salary.std(), size=len(salary))

statistic, p_value = ks_2samp(salary, theoretical_distribution)

print(f"K-S Test Statistic: {statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value > alpha:
    print("The data appears to be consistent with the theoretical distribution (fail to reject the null hypothesis)")
else:
    print("The data does not appear to be consistent with the theoretical distribution (reject the null hypothesis)")
```

## Analysis of important Categorical columns
### 1. Designation

```python
# Writing a function for correcting the spelling error of the values in the categorial columns or reducing the values in columns

from fuzzywuzzy import process

def correct_spelling_errors(target_word="", choices=[], threshold=80):
    match, score = process.extractOne(target_word, choices)
    if score >= threshold:
        return match
    else:
        return target_word
```

```python
non_viz_cat(df["Designation"])
```

```python
sorted(df["Designation"].unique())
```

```python
choices = ['.net developer', 'account manager', 'assistant system engineer', 'associate software engineer', 'assistant manager',
          'business system analyst', 'clerical', 'customer service', 'electrical engineer', 'executive engineer', 'field engineer',
          'front end developer', 'graduate trainee', 'jr. software developer', 'software engineer', 'software tester', 'software trainee',
          'system analyst', 'system administrator', 'system engineer', 'team lead']



df["designation_preprocessed"] = df["Designation"].apply(lambda designation :correct_spelling_errors(designation, choices))
df["designation_preprocessed"].nunique()
```
